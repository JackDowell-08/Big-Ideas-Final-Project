##### The Metric:
The Metric I used was the number of fire incidents in each neighborhood. This is a good metric for comfortability because fire incidents can bring lots of factors which negatively affect comfort, like firetruck sirens, decrease in air quality from smoke, your own house could be more suce, etc.

```python
import pandas as pd

fire = pd.read_csv("fireincidents.csv", index_col='_id')
```

```python
fireplot = fire['neighborhood'].value_counts().sort_values(ascending=True).head(10)
fireplot.plot(kind='bar',xlabel='Neighborhood' , ylabel='Number of Fire Incidents')
```
