```python
import pandas as pd
otp=pd.read_csv("ontimepercent.csv")
#removing any of the buses that count as light rails and routes with unavailable on time percentages
filtered_otp = otp[(otp["mode"]=="Bus") & (otp["on_time_percent"] != 0)]

# Grouping by route, then calculating the average on-time percentage
average_on_time = filtered_otp.groupby(['route_name'])['on_time_percent'].mean().reset_index()

# Renaming the columns
average_on_time.columns = ['Route', 'Avg OTP']

#getting the top 15 routes 
sorted_routes = average_on_time.sort_values(by="Avg OTP", ascending = False)
sorted_routes.head(15)
```

```python
sorted_routes.head(15).plot(kind = "bar", xlabel = "Route", ylabel = "OTP")
```

```python
#getting the bottom 15 routes
sorted_routes.tail(15)
```

```python
sorted_routes.tail(15).plot(kind = "bar", xlabel = "Route", ylabel = "OTP")
```
